```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_percentage_error as mape
```

```python
df = pd.read_csv('./Dummy Data HSS.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
# 결측치 : 각 열의 평균값으로 채우기
print(df.isnull().sum())

df = df.fillna(df.mean())
df.info()
```

### 시각화

```python
num_cols = ['TV', 'Radio', 'Social Media']

fig, axs = plt.subplots(1, 3, figsize = (15, 4))
for i, col in enumerate(num_cols):
    sns.histplot(df[col], kde=True, ax = axs[i])
    axs[i].set_title(f'Histogram of {col}')
    
plt.tight_layout()
plt.show()
```

```python
inf_count = df['Influencer'].value_counts()
inf_count.plot(kind='bar')
plt.title('Influencer count')
plt.show()
```

```python
sns.histplot(df['Sales'], kde=True)
plt.title('Sales histogram')
plt.show()
```

```python
# 명목형 변수 변환
df = pd.get_dummies(df)
df.head(3)
```

```python
df.columns
```

```python
# Sales 변수를 맨 뒤 위치로
df = df[['TV', 'Radio', 'Social Media', 'Influencer_Macro',
       'Influencer_Mega', 'Influencer_Micro', 'Influencer_Nano', 'Sales']]
```

```python
# Sales와의 상관관계 체크
df.corr()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap= 'coolwarm')
```

```python
# 독립, 종속변수 나누기
x = df.iloc[:, 0:-1].values
y = df.iloc[:, -1:].values
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
print(x_train)
print(y_train)
```

```python
print(len(x_train), len(y_train), len(x_test), len(y_test))
```

## 회귀 분석
- 선형회귀
- 랜덤포레스트
- 결정트리
- 서포트벡터
- 다항회귀

### 선형회귀(Linear Regression)

```python
lr_regressor = LinearRegression()
lr_regressor.fit(x_train, y_train)
y_pred_lr = lr_regressor.predict(x_test)

print("r2: ", r2_score(y_test, y_pred_lr))
print("rmse: ", mse(y_test, y_pred_lr)**0.5)
print("mape: ", mape(y_test, y_pred_lr)*100)
```

```python
print(lr_regressor.coef_, lr_regressor.intercept_)
```

```python
df.columns
```

```python
# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_lr, color='blue', label='Data Points')  # 실제값과 예측값 산점도
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Prediction')  # 이상적인 예측 선
plt.title('Actual vs Predicted')  # 그래프 제목
plt.xlabel('True Values')  # x축 레이블
plt.ylabel('Predicted Values')  # y축 레이블
plt.legend()  # 범례 표시
plt.grid(True)  # 그리드 표시
plt.show()  # 그래프 출력
```

### 랜덤포레스트 회귀(Random Forest Regression)

```python
rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train, y_train)
y_pred_rf = rf_regressor.predict(x_test)
```

```python
print("r2: ", r2_score(y_test, y_pred_rf))
print("rmse: ", mse(y_test, y_pred_rf)**0.5)
print("mape: ", mape(y_test, y_pred_rf)*100)
```

```python
# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_rf, color='blue', label='Data Points')  # 실제값과 예측값 산점도
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Prediction')  # 이상적인 예측 선
plt.title('Actual vs Predicted')  # 그래프 제목
plt.xlabel('True Values')  # x축 레이블
plt.ylabel('Predicted Values')  # y축 레이블
plt.legend()  # 범례 표시
plt.grid(True)  # 그리드 표시
plt.show()  # 그래프 출력
```

### 결정트리 회귀(Decision Tree Regressor)

```python
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(x_train, y_train)
y_pred_dt = dt_regressor.predict(x_test)

print("r2: ", r2_score(y_test, y_pred_dt))
print("rmse: ", mse(y_test, y_pred_dt)**0.5)
print("mape: ", mape(y_test, y_pred_dt)*100)
```

### 서포트벡터 회귀(Support Vector Regression)
- 스케일링 필요함

```python
a = df.iloc[:, 0:-1].values
b = df.iloc[:, -1:].values
```

```python
b = b.reshape(len(b), 1)
b
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b)
```

```python
scala = StandardScaler()
scalb = StandardScaler()

a_train = scala.fit_transform(a_train)
b_train = scalb.fit_transform(b_train)
```

```python
sv_regressor = SVR()
sv_regressor.fit(a_train, b_train)
```

```python
b_pred_ = sv_regressor.predict(scala.transform(a_test))
b_pred = b_pred_.reshape(len(b_pred_), 1)

b_pred = scalb.inverse_transform(b_pred)
b_pred
```

```python
print("r2: ", r2_score(b_test, b_pred))
print("rmse: ", mse(b_test, b_pred)**0.5)
print("mape: ", mape(b_test, b_pred)*100)
```

### 다항회귀(Polynomial Regression)

```python
poly = PolynomialFeatures()
po_regressor = LinearRegression()

po_regressor.fit(poly.fit_transform(x_train), y_train)
```

```python
y_pred_po = po_regressor.predict(poly.fit_transform(x_test))
```

```python
print("r2: ", r2_score(y_test, y_pred_po))
print("rmse: ", mse(y_test, y_pred_po)**0.5)
print("mape: ", mape(y_test, y_pred_po)*100)
```

```python
# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_po, color='blue', label='Data Points')  # 실제값과 예측값 산점도
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Prediction')  # 이상적인 예측 선
plt.title('Actual vs Predicted')  # 그래프 제목
plt.xlabel('True Values')  # x축 레이블
plt.ylabel('Predicted Values')  # y축 레이블
plt.legend()  # 범례 표시
plt.grid(True)  # 그리드 표시
plt.show()  # 그래프 출력
```

- 서포트 벡터 회귀(Support Vector Regression)가 가장 높은 R2 점수를 보이고 가장 낮은 RMSE를 가지고 있음 >> 따라서 SVR 모델을 선택
